```python
# LOAD THE AUDIO
```

```python
import librosa
audio_path = "/content/Assignment-1-voice.wav"
x , sr = librosa.load(audio_path)

#x : array
#sr sampling rate in HZ
```

```python
sr
```

```python
x
```

```python
#DISPLAY THE AUDIO
```

```python
import IPython.display as ipd
ipd.Audio(audio_path)
```

```python
import matplotlib.pyplot as plt
import librosa.display
```

```python
#VISUALIZE THE AUDIO
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

```python
#SPECTOGRAM
```

```python
X = librosa.stft(x) #audio to short term fourier transform
Xdb = librosa.amplitude_to_db(abs(X)) #amplitude of frequency at given time
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
import speech_recognition as sr
```

```python
r = sr.Recognizer()
```

```python
#Using record Capture data from the file
sample_audio = sr.AudioFile('Recording_my_voice.wav')
print(type(sample_audio))
with sample_audio as source:
    audio = r.record(source) #duration=3
```

```python
r.recognize_google(audio)
```

```python
#Setting OFFSET
sample_audio = sr.AudioFile('Recording_my_voice.wav')
print(type(sample_audio))
with sample_audio as source:
    audio = r.record(source,duration=3)
```

```python
#Setting OFFSET
sample_audio = sr.AudioFile('Recording_my_voice.wav')
print(type(sample_audio))
with sample_audio as source:
    audio = r.record(source,offset=5,duration=10) #offset : start time, duration : How much length , start from 5sec record for more 10 sec.
```

```python
r.recognize_google(audio)
```
